```python
titles = []
for i in s.find_all('td',class_= 'titleColumn'):
    titles.append(i.text)
```

```python
rating = []
for i in s.find_all('td',class_='ratingColumn imdbRating'):
    rating.append(i.text)
```

9) Write a python program to scrape mentioned details from dineout.co.in :
i) Restaurant name
ii) Cuisine
iii) Location
iv) Ratings
v) Image URL

```python
import pandas as pd 
from bs4 import BeautifulSoup
import requests
```

```python
url = requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
url
```

```python
soup = BeautifulSoup(url.content)
soup
```

```python
r_name = []
for m in soup.find_all('a',class_='restnt-name ellipsis'):
    r_name.append(m.text)
```

```python
cuisine = []
for i in soup.find_all('span',class_='double-line-ellipsis'):
    cuisine.append(i.text.split('|')[1])
```

```python
address = []
for i in soup.find_all('div',class_='restnt-loc ellipsis'):
    address.append(i.text)
```

```python
rating =[]
for n in soup.find_all('div',class_='restnt-rating rating-4'):
    rating.append(n.text)
```

```python
image_link = []
for n in soup.find_all('img',class_='no-img',):
    image_link.append(n['data-src'])
```

```python
image_link
```

```python
restaurent = pd.DataFrame({'Restaurent Name':r_name,'Cuisine':cuisine,'Address':address,'Rating':rating,'Image':image_link})
restaurent.head()
```
